### **Transpose of Matrix**


*   Interchanging rows to columns and vice versa.
*   Matrix transpose is performed with the transpose method on a nested list or a Python array, or a higher-dimensional Numpy array.

**MATRIX** as a **LIST**
* It can be a nested list if there is more than one row.


```python
 # Transpose of a Matrix (as nested list)
 a = [[7,8,2,3,1,4],[5,1,3,4,6,2],[5,7,8,2,1,3]]
 b = np.transpose(a)
 print('a\n',a)
 print('b\n',b)
```

### The **multi_dot** method

* It performs dot (scalar) product with 2 or more input matrices.

* First and last arrays can be either 1D or 2D arrays.
* However, the dimensions of the matrices must suit subsequent scalar matrix multiplication.

**Random module** generates integer between a range say from a to b

```python
 # matrices with random integers: entries ranging from -4 to 4
 a = np.random.randint(-4,4,(30,5))
 b = np.random.randint(-4,4,(5,80))
 c = np.random.randint(-4,4,(80,150))
 d = np.random.randint(-4,4,(150,250))
 e = np.random.randint(-4,4,(250,350))
 # Perform multiple matrix multiplication
 z = np.linalg.multi_dot([a,b,c,d,e])
```

The result of this method can be obtained with successive dot products of matrices but multi_dot functions in an optimized manner. It decides the order of dot multiplication to complete the entire process efficiently.

```python
 %%time
 z = np.linalg.multi_dot([a,b,c,d,e])
 print(z, '\n')
```

```python
 %%time
 z = a.dot(b).dot(c).dot(d).dot(e)
 print(z, '\n')
```

#### Observe the CPU time that multi_dot consumes as against CPU time that successive dot methods consume to arrive at the same solution. Which one is more efficient, why ?

### Inner Product

* The inner product is the scalar multiplication of one vector (or matrix) and the transpose of another vector (or matrix).
* If both arrays are 1D arrays, their dimensions should be identical.
* If either or both arrays are higher-dimensional, then the last dimensions of both arrays should be identical.

```python
 a = np.array([[7,8,2,3],[5,1,3,4],[5,7,8,2]]) # 3 * 4
 b = np.array([[2,3,5,1],[1,5,3,2],[4,6,3,6],[3,1,6,4],[5,3,1,4]]) # 5 * 4
 z = np.inner(a,b) # (3*4) * (4*5)
 print(z)
```

```python
# The same results can be obtained using the dot method as follows.

a.dot(b.T)
```

### Outer Product

* Outer product is the dot product of a column vector of size Mx1 and a row vector of size 1xN.
* The resulting array is a matrix of size MxN.


```python
 a = np.array([5,3,4,9,8])
 b = np.array([3,4,6,8,9,2,1])
 z = np.outer(a,b)
 print(z)
```

* The last dimension of the second array and the second-to-last dimension of the first array should be identical to perform matrix multiplication.
* Further, the symbol @ is also used to perform matrix multiplication.

```python
 # Here, 'a' matrix is 3D, which means there are 3 matrices each of 2x5 size
 # Similarly, for 'b' matrix
 # So we perform 3 matrix multiplication operations each with 2x5 and 5x3 matrices from a and b
 a = np.random.random([4,2,5])
 b = np.random.random([4,5,3])
 z = a @ b
 print(z.shape)
 print(z)
```

### **Matrix Determinant**

### **Matrix Inverse**

* Inverse of a square matrix can be derived using the inv method of the linalg module.
* Dividing the adjoint of a matrix by determinant of a matrix gives the inverse.


```python
 a = np.random.randint(1,20,[3,3])
 inv = np.linalg.inv(a)
 print(a)
 print()
 print(inv)
```

### **Matrix Power**

* Matrix Power is a general method to obtain either positive or negative powers of a given square matrix.
* The first negative power of a matrix is technically termed its inverse.
* Thus, the matrix_power method can be used to find the inverse or any power of a matrix.

```python
 a = np.random.random([3,3])
 # positive powers of matrix
 a_2 = np.linalg.matrix_power(a, 2)
 a_3 = np.linalg.matrix_power(a, 3)
 # inverse of matrix
 a_inv_1 = np.linalg.matrix_power(a, -1)
 a_inv_4 = np.linalg.matrix_power(a, -4)
 print('matrix \n', a)
 print('\n matrix to the power 2\n', a_2)
 print('\n matrix to the power 3\n', a_3)
 print('\n matrix inverse \n', a_inv_1)
 print('\n matrix quadra inverse \n', a_inv_4)
```

### **Eigenvalues and Eigenvectors**

* Eigenvalues and Eigenvectors of a matrix can be determined as follows.
* If Eigen values cannot be determined, the method throws an error (Eg. Singular matrix).
* The vector that changes by a scalar factor after applying a transformation is called an eigen vector, and the scalar value attached to the eigenvector is called the eigenvalue.

```python
 a = np.arange(25).reshape(5,5)
 eig_val, eig_vec = np.linalg.eig(a)
 print('Eigenvalues are: \n', eig_val)
 print('\nEigenvectors are: \n', eig_vec)
```

```python
# Eigenvalues alone can be determined using the method eigvals as shown below.

a = np.arange(25).reshape(5,5)
eigenvalues = np.linalg.eigvals(a)
print(eigenvalues)
```

### Solving System of Equations

* When we think of Linear Algebra, the system of linear equations comes to our mind first, as it is tedious, time-consuming and error-prone.
* NumPy solves systems of linear equations in a fraction of seconds!

```python
# Coefficient Matrix
a = np.random.randint(1,25,[5,5])
# Dependent variable vector
b = np.array([5,8,7,6,3])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
# Check for correctness
B = a.dot(x)
print(B)

# This ‘B’ array is identical to the input ‘b’ array. Hence, our solution is correct.
```

### Singular Value Decomposition

Singular Value Decomposition (SVD) is one of the great dimension-reduction algorithms in machine learning. It identifies the principal components and arranges them by rank. The top ranked components contribute greatly to the original array. Here, we explore SVD with an image to get better understanding.

```python
from skimage import data
# download a sample image
image = data.astronaut()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# Normalize the image by dividing it by the maximum value, 255
# and reorder the axes to be (3,400,600).

# normalize image
img = image/255.0
# reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2,0,1))
```

Perform SVD on the image. It decomposes the original image into three components: U matrix, Sigma vector, and V matrix. The Sigma vector is the diagonal entries of the Sigma matrix. Hence, it is advisable that the Sigma vector may be reformed into a diagonal matrix. It should be noted that the first dimension 3 refers to the three colour channels.

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
# S matrix should have dimensions suitable for matrix multiplication
Sigma = np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(2,1,0))
plt.imshow(reconst)
plt.show()
```

```python
# NumPy SVD reconstruction

# Reconstruct the data by reducing the common dimensions from 512 to 50.

k = 60
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

#### Dimensionality reduction

It is amazing that we reconstructed the image with most details, even at a reduction of ¼.

We can once again reconstruct the same image by reducing the data points from 512 to 20.


```python
k = 30
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

#### SVD reconstruction

Out of 400 data points, merely 20 data points can reconstruct the image with key feature details! This is why the old mathematical algorithm- SVD is still popular in machine learning.
